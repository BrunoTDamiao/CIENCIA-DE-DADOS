#### 2. Faça uma média móvel em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu

````python
df_ma = df['Adj Close'].rolling(window=90).mean()
df_ma
````

#### 2.1 Utilizando a fução rolling(), calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima

````python
df_std = df['Adj Close'].rolling(window=90).std()
df_std
````

#### 2.2 Monte um gráfico de linhas para as médias móveis dos ativos da sua base

````python
df_ma_adjclose = df_ma.stack().reset_index()
df_ma_adjclose.columns = ['data','ativo','valor']

fig = px.line (data_frame=df_ma_adjclose,
               x='data',
               y='valor',
               color='ativo',
               template='plotly_white',
               title='Médias Móveis de Ativos'
              )
fig.show()
````

#### 2.3 Monte um gráfico de linhas para os desvios padrão em janela móvel

````python
df_std_adjclose = df_std.stack().reset_index()
df_std_adjclose.columns = ['data','ativo','valor']

fig = px.line (data_frame=df_std_adjclose,
               x='data',
               y='valor',
               color='ativo',
               template='plotly_white',
               title='Desvio Padrão de Ativos'
              )
fig.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
df_60days = df.xs(key='PANW',
                  level=1,
                  axis=1).tail(n=60)

print(df_60days.shape[0])
df_60days
````

````python
layout = Layout(paper_bgcolor='rgba(0,0,0,0)',
                plot_bgcolor='white',
                height=600)

graph = {
    'x': df_60days.index, 
    'open': df_60days.Open,
    'close': df_60days.Close,
    'high': df_60days.High,
    'low': df_60days.Low,
    'type': 'candlestick',
    'name': 'PANW',
    'showlegend': True
}

grafico = graph_objects.Figure(data=[graph],
                               layout_title='Ativo Palo Alto Networks, Inc.',
                               layout=layout)

grafico.update_xaxes(showgrid=True,
                     gridwidth=1,
                     gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True,
                     gridwidth=1,
                     gridcolor='LightGrey')

grafico
````

- Tendência Geral:
    > Apesar de algumas oscilações, a ação da PANW parece ter se mantido relativamente estável dentro de uma faixa de preço durante o período apresentado. Observamos um pico de alta no final de agosto, seguido por uma leve queda, mas sem grandes rupturas.
