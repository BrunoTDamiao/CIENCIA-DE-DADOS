```python
X = pd.get_dummies(X, drop_first=True)
print(X.info())
X.head()
```

### 1. Execute os passos do exercício anterior, até que você tenha uma árvore de regressão predizendo o valor do imóvel na base de treinamento.

```python
X.corr()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=7547216)
```

```python
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=8)

regr_1.fit(X_train, y_train)
regr_2.fit(X_train, y_train)
```

```python
mse1_train = mean_squared_error(y_true=y_train, y_pred=regr_1.predict(X_train))
mse1_test = mean_squared_error(y_true=y_test, y_pred=regr_1.predict(X_test))

mse2_train = mean_squared_error(y_true=y_train, y_pred=regr_2.predict(X_train))
mse2_test = mean_squared_error(y_true=y_test, y_pred=regr_2.predict(X_test))

template = f'O MSE1 da árvore com profundidade= {regr_1.get_depth()} é: {mse1_test:.2f}\nO MSE2 da árvore com profundidade= {regr_2.get_depth()} é: {mse2_test:.2f}'
print(template)
```

# Arrisco dizer que o MSE2 seria a melhor variavel para se trabalhar 

Esses valores não devem estar corretos, a árvore abaixo nao parece fazer tanto sentindo


```python
plt.rc('figure', figsize=(10,10))
tp = tree.plot_tree(regr_1,
                   feature_names=X.columns,
                   filled=True)
```

### 2.  Calcule o caminho indicado pelos CCP-alfas dessa árvore.

```python
path = regr_2.cost_complexity_pruning_path(X_train, y_train)
path
```

```python
ccp_alphas, impurities = path.ccp_alphas, path.impurities

plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, impurities)
plt.xlabel("Alpha efetivo")
plt.ylabel("Impureza total das folhas")
```

### 3. Paca cada valor de alpha obtido no item 2, treine uma árvore com o respectivo alfa, e guarde essa árvore em uma lista.

```python
clfs = []

for ccp_alpha in ccp_alphas:
    clf = DecisionTreeRegressor(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)
```

```python
tree_depths = [clf.tree_.max_depth for clf in clfs]
plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas[:-1], tree_depths[:-1])
plt.xlabel("effective alpha")
plt.ylabel("Profundidade da árvore")
```

### 4. Para cada árvore na lista, calcule o MSE da árvore.

```python
train_scores = [mean_squared_error(y_train, clf.predict(X_train)) for clf in clfs]
test_scores = [mean_squared_error(y_test, clf.predict(X_test)) for clf in clfs]


```

### 5. Monte um gráfico do MSE pelo alpha, escolha um valor de alpha perto do ponto de mínimo do MSE

```python
fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("MSE")
ax.set_title("MSE x alpha do conjunto de dados de treino e teste")
ax.plot(ccp_alphas[:-1], train_scores[:-1], marker='o', label="treino", drawstyle="steps-post")
ax.plot(ccp_alphas[:-1], test_scores[:-1], marker='o', label="teste", drawstyle="steps-post")
ax.legend()
plt.show()
```

# acho que um valor para ser escolhido seria o valor 0.1.

```python
arvore_final = DecisionTreeRegressor(random_state=0, ccp_alpha=.01, max_depth=10)
arvore_final.fit(X_train, y_train)

print(arvore_final.tree_.max_depth)
```

### 6. Calcule o R-quadrado dessa árvore encontrada no item acima

```python
print(f'o R-quadrado dessa árvore é: {arvore_final.score(X_test, y_test):.2f}')
```
