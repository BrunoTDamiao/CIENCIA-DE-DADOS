### 1. Crie 2 faixas de Latitude do município (munResLat) sendo uma acima e outra abaixo de -10.5 e aplique o groupby usando essas faixas como chave e realize operações de soma, media, minimo, maximo, mediana, desvio padrao, variancia pra pelo menos 2 variáveis numéricas ainda não utilizadas

```python
sinasc_2 = sinasc_raw[['munResLat', 'QTDGESTANT', 'STCESPARTO']].copy()
maior = pd.cut(sinasc_2['munResLat'], [-14, -10.5, -7])
sinasc_2.groupby([maior]).aggregate({'QTDGESTANT' : ['sum','mean','min','max','median'],
                                    'STCESPARTO': ['sum','mean','min','max','median']})
```

### 2. Crie 2 faixas da área dos municípios (munResArea) sendo uma acima e outra abaixo de 3000 e aplique o groupby usando essas faixas como chave e realize operações de soma, media, minimo, maximo, mediana, desvio padrao, variancia pra pelo menos 2 variáveis numéricas ainda não utilizadas


```python
sinasc_3 = sinasc_raw[['munResArea','QTDGESTANT', 'STCESPARTO']].copy()
faixas = pd.cut(sinasc_3['munResArea'], [400, 3000, 35000])
sinasc_3.groupby([faixas]).aggregate({'QTDGESTANT' : ['sum','mean','min','max','median'],
                                    'STCESPARTO': ['sum','mean','min','max','median']})
```

### 3. Determine faixas na variável munResAlt e aplique o groupby usando essas faixas como chave e realize operações de soma, media, minimo, maximo, mediana, desvio padrao, variancia pra pelo menos 2 variáveis numéricas ainda não utilizadas

```python
sinasc_4 = sinasc_raw[['munResAlt','QTDGESTANT','STCESPARTO']].copy()
sinasc_4.value_counts(sort=False)
faixa1 = pd.cut(sinasc_4['munResAlt'], [80,250,350,600])
sinasc_4.groupby([faixa1]).aggregate({'QTDGESTANT' : ['sum','mean','min','max','median'],
                                    'STCESPARTO': ['sum','mean','min','max','median']})
```

### 5. Utilize a tabela do link abaixo e crie faixas utilizando o mapping e gere agrupamentos utilizando essas faixas como chave


```python
dados_ro = pd.read_csv('dados_extras_ro.csv', sep=';', encoding='latin_1')
regioes_map = dict(zip(dados_ro['municipio'], dados_ro['regiao']))
dados_ro.head()
```

### 5.1 IDH
A - https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Rond%C3%B4nia_por_IDH-M


```python
sinasc_6 = sinasc_raw[['munResNome','IDADEMAE']]
sinasc_6= sinasc_6.merge(dados_ro[['idh','ifdm','pib']], left_on='munResNome', right_on=(dados_ro['municipio']))
sinasc_6 = sinasc_6.set_index('munResNome')
result_idh = sinasc_6.groupby(imediatas).agg({'idh':['count','max'], 'IDADEMAE':['count','mean']})
result_ifdm = sinasc_6.groupby(imediatas).agg({'ifdm':['count','max'], 'IDADEMAE':['count','mean']})
result_pib = sinasc_6.groupby(imediatas).agg({'pib':['count','max'], 'IDADEMAE':['count','mean']})
```

```python
fig, axes = plt.subplots(2,2,figsize=(15,12))
dados_ro['IDADEMAE'] = sinasc_raw['IDADEMAE']

x1 = (result_idh['idh']['count'].index)
y1 = (result_idh['idh']['count'].values)
x2 = (result_idh['IDADEMAE']['count'].index)
y2 = (result_idh['IDADEMAE']['count'].values)
x3 = (result_idh['idh']['max'].index)
y3 = (result_idh['idh']['max'].values)
x4 = (result_idh['IDADEMAE']['mean'].index)
y4 = (result_idh['IDADEMAE']['mean'].values)

sns.lineplot(ax= axes[0,0], x=x1, y=y1)
sns.lineplot(ax= axes[0,1], x=x2, y=y2)
sns.lineplot(ax= axes[1,0], x=x3, y=y3)
sns.lineplot(ax= axes[1,1], x=x4, y=y4)


axes[0,0].set_ylabel('IDH - COUNT')
axes[0,0].set_xlabel('REGIÕES')
axes[0,1].set_ylabel('IDADEMAE - COUNT')
axes[0,1].set_xlabel('REGIÕES')
axes[1,0].set_ylabel('IDH - MAX')
axes[1,0].set_xlabel('REGIÕES')
axes[1,1].set_ylabel('IDADEMAE - MEAN')
axes[1,1].set_xlabel('REGIÕES')

plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
