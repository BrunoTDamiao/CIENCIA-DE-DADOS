````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
df.drop_duplicates(inplace=True)
````

````python
df.drop(axis='column', columns='Unnamed: 0', inplace=True)
df.head()
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
plt.figure(figsize=(12, 8))
plt.xticks(rotation=45)


sns.countplot(x='data_ref', hue='educacao', data=df)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

## 1º análise

Analisando o gráfico, podemos perceber que as váriaveis teve algumas alterações consideraveis que poderiam ser investigadas.

- **Ensino Secundário e Superior Completo**:

    * O ensino Secundário, em 2015 nos meses de Janeiro a Agosto estava em alta. Enquanto o Superior completo estava abaixo de 300. Analisando os     meses seguintes podemos dizer que mais pessoas começaram a terminar o ensino superior e consequentemente diminuiu os dados de ensino secundário.

- **Superior Incompleto, Prímario e Pós Graudação**:

    * Ensino Superior Incompleto e primário, não obteve nenhuma grande diferença no decorrer dos meses. Enquanto a Pós graduação podemos analisar que teve aparições minimas nos meses de Março, Maio e Julho de 2015.

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
sns.set(style="darkgrid")


plt.figure(figsize=(12, 8))

sns.pointplot(data=df, x='data_ref', y='renda', hue='sexo', dodge=True)

plt.title('Média da Coluna Sexo por Nível da Variável Renda ao Longo do Tempo')

plt.xlabel('Tempo')

plt.ylabel('Média de Posse de Veiculo')

plt.legend(title='Sexo', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.xticks(rotation=45)

plt.show()
````

Bom, analisando o gráfico acima, podemos dizer que teve poucas variações entre o sexo F e M.
